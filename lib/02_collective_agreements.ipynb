# Collective Agreements

The ideal experiment would be one in which probationary period length is randomly assigned to firms. Unfortunately, this is not feasible. However, one may attempt to obtain the causal effects of probationary period length on separations by controlling for confounders. Such an approach could exploit the fact that collective agreements signed between unions and firms generally stipulate the length of probationary periods.

Taking the set matches under a collective agreement as the unit of observation, one can come up with a causal graph for the histogram of separations by tenure $(Y)$. The figure below is a naive causal graph that ignores variation across occupation-geography-time.

The causal effect of probationary period length $(PP)$ on the histogram of separations by tenure $(Y)$ is identified without any conditioning. However, occupation-geography-time generate backdoor paths since these variables have a direct impact on:
* $Y$, $QA$, and $QH$ through shocks in the economy affecting regions and occupations differently over time
* $PP$ through the collective agreement which usually applies to a specific (occupation-geography-time) triple
* $US$ through historical and economic circumstances of organized labor for an occupation in a region
* $LD$ through the fact that the scope of learning-by-doing is likely to vary by occupation

Hence, one could simplify the above causal graph by setting:
* $Z_1=US$
* $Z_2=\{RP,MI,LD,QA,QH\}$
* $Z_3=\{UI,MM\}$
* $S=$ occupation-state-year

### Discussion

* The fact that occupation-state-year ($S$) creates backdoor paths need not be too problematic. In fact controlling for these observed variables blocks all backdoor paths.
* However, the issue is that collective agreements are usually defined at the occupation-state-year level. Hence, there will be little variation in $PP$ for some given $S$, making it very complicated to estimate via matching procedures.
* Therefore, a way forward would be to exploit some sort of quasi-experimental variation in the observational data. The clearest example is "bunching" in separations at the end of probationary periods.
* Such an approach requires a model that looks into firms' decisions to terminate matches. Its corresponding causal graph will essentially go into the economic behavior of how individual separations aggregate up to $Y$.
* The usefulness of the previous causal graph is that it informs us that estimates for each occupation-state-year triple should be separate. In addition, counterfactual simulations will be assuming that $RP$, $MI$, and $QH$ remain constant.
