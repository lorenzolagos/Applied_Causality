# Motivation

Firing-hiring rules affect the functioning of the labor market. Job security legislation is one of the main tools that policymakers have to affect such rules. This type of legislation determines (among other things):
* Severance: pecuniary costs of employer paid to the employee upon layoff
* Firing fines: pecuniary costs of employer paid to the government upon layoff
* Probationary periods: tenure cutoff below which lower/no pecuniary costs of layoffs apply

Lower severance and firing fines benefit employers since they reduce the cost of layoffs. The effects on workers are ambiguous because employees would be more likely to lose their jobs (potentially with lower payouts) but such policies may also increase the probability of finding a new job as firms increase hiring. Probationary periods are a sort of compromise. On one hand, they allow employers to test workers at the workplace without the burden of layoff costs for a fixed period of time, thereby increasing the likelihood for potential workers of becoming employed. On the other hand, once the worker's tenure crosses the probationary period cutoff, all regular severance and firing costs apply so that workers have some job security.

Probationary periods are particularly interesting because they introduce quasi-experimental variation in observational data. That is, compared to a world without probationary periods, one would expect a large mass of separations before the end of the probationary period and a missing mass of separations after the end of the probationary period. These empirical moments are quasi-experiental in the sense that there are workers of marginally different abilities whose survival past the probationary period is distinct because of the discontinuity in firing costs. Hence, the parameters of an economic model can be chosen so as to minimize the difference between these moments and simulated data. Finally, the estimated model allows us to generate any desired counterfactual, thereby providing answers to questions such as:
* What is the percentage change in separations at the end of the probationary period in response to a percentage increase in firing costs (i.e., firing cost elasticity of separations)?
* How does the firing cost elasticity of separations vary by probationary period length, occupation, time, region, etc?
* What is the effect of changing the probationary period length on the percentage of workers not surviving past the probationary cutoff (i.e., probationary turnover)?
* Which probationary period length minimizes probationary turnover? 

### The Brazilian Case

In Brazil, the law stipulates the following regarding probationary contracts:
* Maximum 90 day period
* Contract can be renewed once without the sum exceeding 90 days
* If layoff occurs before the end of the probationary period, severance is comprised of half of the remaining wages

The image below illustrates the firing cost schedule that an employer would face for a worker under a 90 day probationary contract. Note that at the end of the probationary period there is a discontinuity in both slope and levels. These discontinuities are what lead to bunching in separations.
